```python
###############
### Imports ###
###############

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import gridspec            # to arrange the plots nicely
%matplotlib inline

'''
params = {
    'font.family': 'serif',
    etc.
}
plt.rcParams.update(params)
'''
plt.close()

import ROOT
```

```python
#def get_tree(filename):
#    f = ROOT.TFile(filename)
#    print(f)
#    f.cd('events')
#    print(f)
#    tree = f.Get('events').Get('events')
#    print(tree)
#    return tree

def get_position(filename):
    f = ROOT.TFile(filename)
    f.cd('events')
    tree = f.Get('events').Get('events')
    x=[]
    y=[]
    z=[]
    r=[]
    for entry in tree:
        x.append(entry.xp_pri)
        y.append(entry.yp_pri)
        z.append(entry.zp_pri)
        r.append(np.sqrt(entry.xp_pri*entry.xp_pri+entry.yp_pri*entry.yp_pri))
    return r, z, x, y

def plot(color, label, alpha=0.9):
    ax.scatter(r,z, color=color, edgecolor='none', marker='s', s=3, alpha=alpha, label=label)
    colors.append(color)
    return

def plottop(color, label, alpha=0.9):
    ax.scatter(x,y, color=color, edgecolor='none', marker='s', s=3, alpha=alpha, label=label)
    colors.append(color)
    return

def plotside(color, label, alpha=0.9):
    ax.scatter(x,z, color=color, edgecolor='none', marker='s', s=3, alpha=alpha, label=label)
    colors.append(color)
    return

def plotyz(color, label, alpha=0.9):
    ax.scatter(y,z, color=color, edgecolor='none', marker='s', s=3, alpha=alpha, label=label)
    colors.append(color)
    return
```
